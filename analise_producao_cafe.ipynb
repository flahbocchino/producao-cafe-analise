```r
install.packages("readxl")
library(readxl)

# Lendo o arquivo com nome exato do seu upload
dados <- read_excel("/content/producao_cafe.xlsx.xlsx")

# Mostrar os dados
head(dados)
```

```r
# Tabela de frequência
tabela <- table(dados$`Tipo de Cultivo (Nominal)`)
print(tabela)

```

```r
barplot(tabela,
        main = "Distribuição dos Tipos de Cultivo de Café",
        xlab = "Tipo de Cultivo",
        ylab = "Frequência",
        col = c("darkgreen", "saddlebrown"))
```

```r
pie(tabela,
    main = "Proporção dos Tipos de Cultivo de Café",
    col = c("darkgreen", "saddlebrown"))
```

```r
# Medidas de tendência central
media <- mean(dados$`Produtividade (sacas/ha) (Contínua)`)
mediana <- median(dados$`Produtividade (sacas/ha) (Contínua)`)

# Função para calcular a moda
moda <- function(v) {
  uniqv <- unique(v)
  uniqv[which.max(tabulate(match(v, uniqv)))]
}
moda_valor <- moda(dados$`Produtividade (sacas/ha) (Contínua)`)

# Medidas de dispersão
desvio_padrao <- sd(dados$`Produtividade (sacas/ha) (Contínua)`)
variancia <- var(dados$`Produtividade (sacas/ha) (Contínua)`)
amplitude <- range(dados$`Produtividade (sacas/ha) (Contínua)`)

# Medidas separatrizes (quartis)
quartis <- quantile(dados$`Produtividade (sacas/ha) (Contínua)`, probs = c(0.25, 0.5, 0.75))

# Exibir resultados
cat("Média:", media, "\n")
cat("Mediana:", mediana, "\n")
cat("Moda:", moda_valor, "\n")
cat("Desvio padrão:", desvio_padrao, "\n")
cat("Variância:", variancia, "\n")
cat("Amplitude:", amplitude[1], "a", amplitude[2], "\n")
cat("1º Quartil (25%):", quartis[1], "\n")
cat("2º Quartil (50% - mediana):", quartis[2], "\n")
cat("3º Quartil (75%):", quartis[3], "\n")
```
